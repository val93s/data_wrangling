### Part 1: Merge df1 and df2

```python
df1 = pd.DataFrame({
    "bookingID" : [15926, 17932, 97845, 35784, 17932],
    "resDate" : ["Dec 11", "Dec 11", "Nov 15", "Dec 11", "Oct 09"]
})
df1
```

```python
df2 = pd.DataFrame({
    "bookingID" : [15926, 35784, 17932],
    "clientName" : ["Flynn. Elisse", "Nedresky, Kayla" , "Segall, Alison"],
    "roomType" : ["Lake", "Garden", "Park"],
})
df2
```

Fill in the merge below. What are the duplicates?

#### Part 3: Outer Join

To make `df5`, join df1 and df2 on `"bookingID"` and set `how = "outer"`.

```python
df5 = pd.merge(df1,df2, on = 'bookingID',how ='outer')
df5
```

```python
df6 = pd.merge(df1,df2_new, on = 'bookingID',how ='right')
df6
```

The missing values have occurred because we have joined DataFrames that do not contain the exact same data. Since we specified an outer join, we are including everything from both DataFrames where booking ID is similar. If it is not, a new row is created despite missing values from the column.
